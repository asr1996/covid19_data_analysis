```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Project.csv')
df.index = np.arange(1,len(df)+1)
df.head(10)
```

```python
df.info()
```

```python
df1=df.fillna(0)
df1
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df_sorted_pop=df1.sort_values(by=["Population"],ascending=False)
df_sorted_pop.head(10)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
top10_confirmed = pd.DataFrame(df1.groupby('Country')['Population'].sum().nlargest(10).sort_values(ascending = False))
fig1 = px.scatter(top10_confirmed, x = top10_confirmed.index, y = 'Population', size = 'Population', size_max = 120,
                color = top10_confirmed.index, title = 'World Population')
fig1.show()
```

```python
df_sorted_death=df1.sort_values(by=["Total Deaths"],ascending=False)
df_sorted_death.head(10)
```

```python
top10_deaths = pd.DataFrame(df.groupby('Country')['Total Deaths'].sum().nlargest(10).sort_values(ascending = True))
fig2 = px.bar(top10_deaths, x = 'Total Deaths', y = top10_deaths.index, height = 600, color = 'Total Deaths', orientation = 'h', template='plotly_dark',
            color_continuous_scale = ['deepskyblue','red'], title = 'Top 10 Countries with Maximum Death Count')
fig2.show()
```

```python
df_sorted_recovered=df1.sort_values(by=["Total Recovered"],ascending=False)
df_sorted_recovered.head(10)
```

```python
top10_recovered = pd.DataFrame(df.groupby('Country')['Total Recovered'].sum().nlargest(10).sort_values(ascending = False))
fig3 = px.bar(top10_recovered, x = top10_recovered.index, y = 'Total Recovered', height = 600, color = 'Total Recovered',
             title = 'Top 10 Countries with Recovered Cases', color_continuous_scale = px.colors.sequential.Viridis)
fig3.show()
```

```python
world = df1.groupby('Country')['Population','Active Cases','Total Cases','Total Deaths','Total Recovered'].sum().reset_index()
fig4 = px.choropleth(world, locations ='Country', 
                    locationmode = 'country names',color = 'Population',
                    range_color = [1,400000000]
                   )
fig4.show()
```

```python
fig5 = px.choropleth(world, locations ='Country', 
                    locationmode = 'country names',color = 'Total Cases',
                    range_color = [1,100000000]
                   )
fig5.show()
```

```python
fig6 = px.choropleth(world, locations ='Country', 
                    locationmode = 'country names',color = 'Total Deaths',
                    range_color = [1,1200000]
                   )
fig6.show()
```

```python
fig7 = px.choropleth(world, locations ='Country', 
                    locationmode = 'country names',color = 'Total Recovered',
                    range_color = [1,100000000]
                   )
fig7.show()
```

```python
df_sorted_cases=df1.sort_values(by=["Total Cases"],ascending=False)
fig8=px.bar(df_sorted_cases.iloc[0:20][::-1],
      y='Country',
      x=['Total Cases','Total Recovered', 'Active Cases','Total Deaths','Serious/Critical'], template='plotly_dark',
      title='Severly Hit Countries with Recovery')
fig8.update_xaxes(tickangle=0)
fig8.show()
```

```python
px.pie(df_sorted_pop.iloc[0:20],
       names='Country',
       values='Population',
      template='plotly_dark',
      title='Distribution of Population of Top 20 countries')
```

```python
px.pie(df_sorted_cases.iloc[0:20],
       names='Country',
       values='Total Cases', 
       template='plotly_dark',
      title='Distribution of Total Cases of top 20 Countries')
```

```python
fig11 = px.scatter_3d(df_sorted_cases.head(20), x='Total Cases', y='Total Deaths', z='Total Recovered',
              color='Country')
fig11.update_layout(title='3D Plot of Total Cases, Total Deaths and Total Recovered Cases of Top 20 effected Countries')
fig11.show()
```

```python
df_sorted_pop1=df_sorted_pop[0:30]
px.line(df_sorted_pop1,
       x='Country',
       y=['Total Cases', 'Total Deaths','Total Recovered', 'Active Cases'],
       template='plotly_dark',
       title='Country wise trends of Covid-19 cases',
       labels={'value':'Population'})
```

```python
df_sorted_pop.iloc[0:20].plot(x="Country", y=['Population',"Total Cases"], 
                                kind="bar", figsize=(20,10), width=0.9,xlabel='Top 20 Countries With Highest Population',
                                ylabel='Count in Billions')
```

```python
df_sorted_cases.iloc[0:20].plot(x="Country", y=["Total Cases","Total Recovered","Total Deaths"], 
                                kind="bar", figsize=(20,10), width=0.9,xlabel='Top 20 Countries With Highest Cases',
                                ylabel='Count per 100 Millions')
```
